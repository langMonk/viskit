```python
import time

import pandas as pd
import umap.umap_ as umap
from memory_profiler import memory_usage
from sklearn.manifold import TSNE

dataset_name = "smallnorb"

dataset = pd.read_csv('/Users/bartoszminch/Documents/Repositories/dataset_viskit/datasets/{}_data.csv'.format(dataset_name), header=None)
labels = pd.read_csv('/Users/bartoszminch/Documents/Repositories/dataset_viskit/datasets/{}_labels.csv'.format(dataset_name), header=None)
```

```python
umap = umap.UMAP()
def f():
    return umap.fit_transform(dataset)

umap_start = time.time()
mem_usage = memory_usage(f)
umap_end = time.time()
```

```python
print("Casting time [s]: {}".format(umap_end - umap_start))
print('Maximum memory usage [MB]: {}'.format(max(mem_usage) - min(mem_usage)))
```
