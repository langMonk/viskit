```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


def draw_2d_sns(X, Y):
    df = pd.DataFrame()
    df["X"] = X
    df["Y"] = Y

    plt.figure(figsize=(16, 16))

    sns.scatterplot(
        x="X",
        y="Y",
        data=df,
        legend="full",
        alpha=0.8,
        edgecolor=None,
    )

    plt.legend(loc="right", markerscale=3.0, fontsize=16)
```

```python
import umap
from viskit.embedding.ivhd import Ivhd

X = pd.read_csv("./mammoth.csv", header=None, index_col=None)

mapper = umap.UMAP()
embedding = mapper.fit_transform(X)

draw_2d_sns(embedding[:,0], embedding[:,1])
plt.show()

# mapper = Ivhd(
#     optimizer="force-directed",
#     n_iter=2000,
#     nn=2,
#     rn=1,
#     l1_steps=0,
#     reverse_neighbors_steps=0,
#     reverse_neighbors_count=0,
# )
#
# embedding, _ = mapper.fit_transform(X)
# draw_2d_sns(embedding[:,0], embedding[:,1])
# plt.show()
```
