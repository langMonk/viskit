# Visualizations using pyviskit

```python
from pyviskit import IVHD, types
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def visualize(ax, df, optimizer):
    classes = [i for i in range(10)]

    for cls in classes:
        indices = df.loc[(df['class'] == cls)]
        ax.scatter(indices['x'], indices['y'], label=f"{cls}", s=5)

    ax.legend()
    ax.set_title(f"Mnist for {optimizer}")
```

```python
optimizers = [types.OptimizerType.ForceDirected, types.OptimizerType.Momentum, types.OptimizerType.Adadelta, types.OptimizerType.Adam, types.OptimizerType.Nesterov]
ivhds = [IVHD.IVHD(optimizer, iterations=5000) for optimizer in optimizers]
results = []
optimizers_names = ["Force Directed", "Momentum", "Adadelta", "Adam", "Nesterov"]
```

```python
results.append(ivhds[0].fit_transform('../../../../viskit_offline/mnist_data.csv', '../../../../viskit_offline/mnist_targets_2.csv', '../../../../viskit_offline/mnist_euclidean.bin'))
```

```python
fig, axes = plt.subplots(3, 2)
fig.set_figwidth(10)
fig.set_figheight(15)

for ax, df, optimizer_name in zip(axes.flatten()[:-1], results, optimizers_names):
    visualize(ax, df, optimizer_name)
```
