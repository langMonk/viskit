# Example usage of pyviskit library

```python
from pyviskit import types, IVHD
import pandas as pd
import matplotlib.pyplot as plt
```

## Create instance of caster

Be sure to use (caster, optimizer) pair that is defined (read pyviskit library README). Otherwise, you'll get an error which will tell you, what you did wrong.

```python
# optimizer type
optimizer = types.OptimizerType.ForceDirected
```

```python
# caster
IVHD = IVHD.IVHD(optimizer)
```

## Fit Transform

```python
input_data = "../../mnist_data.csv"
labels = "../../mnist_targets_2.csv"
graph_file = "../../mnist_euclidean.bin"

df = IVHD.fit_transform(input_data, labels, graph_file)
```

```python
df.shape
```

```python
df.head(10)
```

```python
for target in df['class'].unique():
    indices_to_keep = df.loc[df['class'] == target]
    plt.scatter(indices_to_keep['x'], indices_to_keep['y'], label=f"{target}", s=10)

plt.legend()
plt.show()
```
